```python
import pandas as pd
```

```python

file_path1 = '/content/Sales_January_2019.csv'
file_path2 = '/content/Sales_February_2019.csv'
file_path3 = '/content/Sales_March_2019.csv'
file_path4 = '/content/Sales_April_2019.csv'
file_path5 = '/content/Sales_May_2019.csv'
file_path6 = '/content/Sales_June_2019.csv'
file_path7 = '/content/Sales_July_2019.csv'
file_path8 = '/content/Sales_August_2019.csv'
file_path9 = '/content/Sales_September_2019.csv'
file_path10 = '/content/Sales_October_2019.csv'
file_path11 = '/content/Sales_November_2019.csv'
file_path12 = '/content/Sales_December_2019.csv'
```

```python
## Load the data into seperate dataframes
df_month1 = pd.read_csv(file_path1)
df_month2 = pd.read_csv(file_path2)
df_month3 = pd.read_csv(file_path3)
df_month4 = pd.read_csv(file_path4)
df_month5 = pd.read_csv(file_path5)
df_month6 = pd.read_csv(file_path6)
df_month7 = pd.read_csv(file_path7)
df_month8 = pd.read_csv(file_path8)
df_month9 = pd.read_csv(file_path9)
df_month10 = pd.read_csv(file_path10)
df_month11 = pd.read_csv(file_path11)
df_month12 = pd.read_csv(file_path12)
```

```python
combined_df = pd.concat([df_month1, df_month2, df_month3,df_month4, df_month5, df_month6, df_month7 ,df_month8, df_month9, df_month10, df_month11, df_month12], ignore_index=True)
```
